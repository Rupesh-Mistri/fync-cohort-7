```python
import pandas as pd
 
# Creating the DataFrame
df = pd.DataFrame({'Weight': [45, 88, 56, 15, 71],
                   'Name': ['Sam', 'Andrea', 'Alex', 'Robin', 'Kia'],
                   'Age': [14, 25, 55, 8, 21]})
 
# Create the index
index_ = ['Row_1', 'Row_2', 'Row_3', 'Row_4', 'Row_5']
df.index = index_
print(df)
```

### Using `loc` in pandas

```python
result = df.loc[['Row_2', 'Row_3'] , 'Name']
print(result)
```

### Stack plot

```python
# plt.style.use('Solarize_Light2')
experience = [1,3,4,5,7,8,10,12]
salary1 = [6500, 9280, 12050, 13200, 16672, 21000, 23965, 29793]
plt.title("Salary of employee by their experience")
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.stackplot(experience, salary1, colors = 'g')
plt.show()
```

```python
plt.style.use("fivethirtyeight")
plt.rcParams['font.size'] = 10
days = [1,2,3,4,5,6]

sleep = [6,7,5,8,6,7]
drinking_water = [2,2,1,2,1,1]
work =  [5,7,10,8,6,9]
exercise=  [3,3,0,1,3,2]
plt.plot([], [], color = "green", label = 'sleep', linewidth=3)
plt.plot([], [], color = "blue", label = 'drinking_water', linewidth=3)
plt.plot([], [], color = "red", label = 'work', linewidth=3)
plt.plot([], [], color = "orange", label = 'exercise', linewidth=3)

plt.xlabel('days')
plt.ylabel('activities')
plt.title('6 days routine stack plot')
plt.stackplot(days, sleep, drinking_water, work, exercise, colors=['green', 'blue', 'red', 'orange'])
plt.legend(loc="upper right")
plt.show()
```

```python
stock1= [5,3,3,6,1,8,2,7,9]
stock2= [2,4,1,3,5,0,3,1,0]
stock3= [2,2,5,1,3,1,4,1,0]

days =[1,2,3,4,5,6,7,8,9]

stocks= ["stock1","stock2","stock3"]
colors = ["#F9CDAD", "#FC9D9A", "#83AF9B"]

plt.title("Stack Plot of Stock Rates")

plt.stackplot(days,stock1,stock2,stock3,labels=stocks,colors=colors)

plt.legend()
plt.tight_layout()
plt.show()
```
